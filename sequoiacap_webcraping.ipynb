```python
# Install the libraries
import time
from getpass import getpass
import os # for interacting with the operating system
import pandas as pd

# library for directory location:
import pathlib
from os.path import join

# Install Classes to interacting with the web page
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
#Initiate the web driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# Open the website
driver.get("https://www.sequoiacap.com/our-companies/?_categories=ai-ml#all-panel")
time.sleep(2)
```

```python
# Accept the cookie policy
accept_button_xpath = "//button[@class='button button--color-mode button--bg button--tiny js-cookie-accept']"
accept_button = driver.find_element(By.XPATH,accept_button_xpath)
accept_button.click()
time.sleep(2)
```

```python
# go to the end of the page in order for all the elements to be loaded
load_more_xpath = '//*[@id="all-panel"]/div[2]/button'
load_more = driver.find_element(By.XPATH, load_more_xpath)
time.sleep(1)
# Scroll to the element
driver.execute_script("arguments[0].scrollIntoView();", load_more)
time.sleep(2)
load_more.click()
time.sleep(2)
```

```python
#find all <tr> elements with the attribute aria-expanded set to "false" and data-toggle set to "collapse," 
app_list = driver.find_elements(By.CSS_SELECTOR, 'tr[aria-expanded="false"][data-toggle="collapse"][data-target^="#company_listing-"][data-aos="fade-in"][data-aos-duration="1000"][data-aos-delay="100"].aos-init.aos-animate')
```

```python

# First create an empty list for the companies to append the information from each company
company_info = []

# Miximaze the window
try:
    # Check if the window is maximized before attempting to maximize it
    if driver.get_window_rect()['width'] == driver.execute_script("return window.screen.availWidth") and \
            driver.get_window_rect()['height'] == driver.execute_script("return window.screen.availHeight"):
        print("Window is already maximized.")
    else:
        # Scroll to the top of the page
        driver.execute_script("window.scrollTo(0, 0);")
        # Maximize the window
        driver.maximize_window()
        time.sleep(2)  
except Exception as e:
    print(f"An error occurred: {str(e)}")

# Scroll to the top of the page
driver.execute_script("window.scrollTo(0, 0);")

time.sleep(2)
# Loop through the app_list
i = 0
j = 0
for app in app_list:
    # Scroll to the element to make it clickable
    driver.execute_script("arguments[0].scrollIntoView();", app)
    time.sleep(1)
    company_name = app.find_element(By.CSS_SELECTOR, 'th[scope="row"].company-listing__cell-wide.company-listing__head').text
    time.sleep(1)
    current_stage = app.find_element(By.CSS_SELECTOR, 'td.u-lg-hide').text
    time.sleep(1)
    try:
        # Click the element
        app.click()
        time.sleep(2)  # Wait for the page to load

        # Find the elements in the introduction
        company_intro = driver.find_elements(By.CSS_SELECTOR, '.wysiwyg.wysiwyg--fs-lg')[j].text
        j=j+1
        time.sleep(1)
        # Find the links
        company_links = driver.find_elements(By.CLASS_NAME, "l-pillbox.l-hr-row__item")[i]
        company_official_web = company_links.find_element(By.TAG_NAME, 'a').get_attribute("href")
        time.sleep(1)
        category_elements = driver.find_elements(By.CLASS_NAME, "l-pillbox.l-hr-row__item")[i+1].find_elements(By.CSS_SELECTOR, 'a')
        company_category =[]
        for element in category_elements:
            company_category.append(element.text)
        i = i+2
        time.sleep(1)

        # Populate the list
        columns = [company_name, current_stage, company_category, company_intro, company_official_web]
        company_info.append(columns)
        time.sleep(2)
          

    except Exception as e:
        print(f"Error clicking element: {e}")    

    
```

```python
# Close the browser window
driver.quit()
```
