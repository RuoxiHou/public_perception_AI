```python
import pandas as pd
```

```python
qualtrics_data = pd.read_excel("Qualtrics_All countries.xlsx")
qualtrics_data.head()
```

```python
qualtrics_data.info()
```

```python
surveymonkey_data = pd.read_excel('SurveyMonkey_America.xlsx')
surveymonkey_data.head()
```

```python
surveymonkey_data.info()
```

```python
#All the data types for the columns are the same for both dataframe. 
#There are no missing values from either dataframe benefiting from the survey design.
# However, the scale of the rating from Survey Monkey is 10 times bigger than the one from Qualtrics and the starting point is 0.
#Therefore, the rating values from Qualtrics and SurveyMoney need to be unified
# Before doing the scale change, the names of the dataframe need to be adjusted for the practical use as the names are too long
column_name = ['id', 'start_time', 'end_time', 'country', 'age', 'sex', 'education', 'knowledge', 'general', 'customer_relationship', 'healthcare', 'self_driving_tech', 'law_firm']
# Use a dictionary to map the old column names to the new ones
column_name_mapping = dict(zip(qualtrics_data.columns, column_name))
qualtrics_data.rename(columns=column_name_mapping, inplace=True)
qualtrics_data.head()
```

```python
#Same renaming needs to be done on surveymonkey_data
column_name_mapping = dict(zip(surveymonkey_data.columns, column_name))
surveymonkey_data.rename(columns=column_name_mapping, inplace=True)
surveymonkey_data.head()
```

```python
columns_to_apply =['knowledge', 'general', 'customer_relationship', 'healthcare', 'self_driving_tech', 'law_firm']
```

```python
# First multiply the numerical columns by 10 in qualtrics_data
qualtrics_data[columns_to_apply] = qualtrics_data[columns_to_apply]*10
```

```python
qualtrics_data.head()
```

```python
# Define the min_valuea function to transform the min data in surveymondy_data into 10
def min_value(value):
    if value <= 10:
        return 10
    else:
        return value
```

```python
surveymonkey_data[columns_to_apply] = surveymonkey_data[columns_to_apply].applymap(min_value)
```

```python
survey_data = pd.concat([qualtrics_data, surveymonkey_data])
```

```python
survey_data.describe()
```

```python
#We change the education into shorter words

# Define a mapping of current values to new values
value_mapping = {"Bachelor’s or the equivalent": 'Bachelor', "Master’s or the equivalent": "Master", "High school graduate, diploma or the equivalent": 'High school', 'Doctoral or the equivalent': 'Doctor', 'Higher than Doctoral': 'Above Doctor'}

# Replace the values in the 'education' column
survey_data['education'] = survey_data['education'].replace(value_mapping)

survey_data['education'].value_counts()
```

```python
#Since there are not many Doctor or Above Doctor, we combine the two together
survey_data['education'] = survey_data['education'].replace({'Doctor': 'Doctor and above', 'Above Doctor': 'Doctor and above'})
survey_data['education'].value_counts()
```
