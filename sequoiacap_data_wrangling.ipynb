```python
# Because the company might fall into different categories, it's better to have the category names splitted into e.g. "category_1", "category_2" 
# First we check how the category column looks like
company_info_df['category'].value_counts()
```

```python
#According to the description of the companies, the category is not suitable for the need of this study
#Based on the survey design, the categorization of the companies is changed into 2 levels

#1st level: Enterprise(product is designed for enterprise directly), Consumer(product is designed for individual consumers directly), AI/ML(product's target audience is not defined)
#2nd level: Customer relationship, Healthcare, Self-driving technology, Law firm, and Others
```

```python
'''All the companies categorized as "CONSUMER" are moved into the new 1st category "Consumer";
Same goes for "ENTERPRISE" to "Enterprise";
The rest goest to AI/ML in the 1st level;
All the companies categorized as "HEALTHCARE" are moved into the new category "Healthcare";
Use keywords for the categorization of "Customer relationship", "Self-driving technology" and "Law firm"
 '''
```

```python
#First change the intro text in lower case
company_info_df['intro'] = company_info_df['intro'].str.lower()
```

```python
#Same applies to the column category
company_info_df['category'] = company_info_df['category'].str.lower()
```

```python
#define a function to create an extra column for the 1st-level category
def category_1(old_category):
    if "consumer" in old_category:
        return "Consumer"
    elif "enterprise" in old_category:
        return "Enterprise"
    else:
        return "AI/ML"

```

```python
company_info_df['category1'] = company_info_df['category'].apply(category_1)
```

```python
def category_2(old_category, intro_text):
    customer_manage_list = ['customer', 'contact center', 'chatbot', 'customer service']
    self_driving_list = ['self-driving','vehicle', 'fleet']
    law_firm_list =['legal', 'law', 'contract']
    if any(kw in old_category or kw in intro_text for kw in customer_manage_list):
        return "Customer relationship"
    elif 'healthcare'in old_category:
        return "Healthcare"
    elif any(kw in old_category or kw in intro_text for kw in self_driving_list):
        return "Self-driving technology"
    elif any(kw in old_category or kw in intro_text for kw in law_firm_list):
        return "Law firm"
    else:
        return "Other"
```

```python
company_info_df['category2'] = company_info_df.apply(lambda row: category_2(row['category'], row['intro']), axis=1)
```
